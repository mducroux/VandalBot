```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("output.txt", sep="\t", header=None)
data.columns = ["text", "label"]
data["text"] = data["text"].apply(lambda x: str(x)[2:])
data["label"] = data["label"].apply(lambda x: str(x)[:-1])
```

```python
data.to_csv("input.txt", sep="\t", header=None, index=None)
```
